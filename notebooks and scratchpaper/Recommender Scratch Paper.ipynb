```python
temp = []
#Output the pairs with their score
for i in range(len(searched_embeds)):
    #print("{} \t\t {} \t\t Score: {:.4f}".format(searched_descriptions[i], liked_descriptions[i], cosine_scores[i][i]))
    max_index = np.argmax(cosine_scores[i])
    print(f"{i} \t {df['title'][i]} \nMax Score at {max_index}: {cosine_scores[i][max_index]}\n")
    temp.append({i: float(cosine_scores[i][max_index])})
```
